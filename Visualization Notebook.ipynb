```python
import pandas as pd 
import numpy as np                     
import seaborn as sns                  
import matplotlib.pyplot as plt        
%matplotlib inline 
import warnings       
import scipy as sp
plt.rcParams.update(plt.rcParamsDefault)

plt.rcParams['font.family'] = "serif"
```

# Students

```python
%cd /Users/mayanksharma/Desktop/T4P/Students
```

```python
data=pd.read_csv("analysis.csv")
```

```python
plt.rcParams["figure.figsize"] = (12,4)

pre_means = (np.mean(data['presec']),np.mean(data['preseo']),np.mean(data['presc']),np.mean(data['precogreapp']),np.mean(data['preexpsupp']),np.mean(data['prerational']),np.mean(data['preintuitive']),np.mean(data['preempathy']))
post_means = (np.mean(data['postsec']),np.mean(data['postseo']),np.mean(data['postsc']),np.mean(data['postcogreapp']),np.mean(data['postexpsupp']),np.mean(data['postrational']),np.mean(data['postintuitive']),np.mean(data['postempathy']))

pre_std= (np.std(data['presec']),np.std(data['preseo']),np.std(data['presc']),np.std(data['precogreapp']),np.std(data['preexpsupp']),np.std(data['prerational']),np.std(data['preintuitive']),np.std(data['preempathy']))
post_std= (np.std(data['postsec']),np.std(data['postseo']),np.std(data['postsc']),np.std(data['postcogreapp']),np.std(data['postexpsupp']),np.std(data['postrational']),np.std(data['postintuitive']),np.std(data['postempathy']))

patterns = [ "|" , "\\" , "/" , "+" , "-", ".", "*","x", "o", "O" ]

r = np.array([0,0.5,1,1.5,2,2.5,3,3.5])
width=0.15


plt.bar(r, pre_means, width, color='gold', yerr=pre_std,edgecolor='k', capsize=7, label='Pre')
plt.bar(r + width, post_means,width, color='darkorange', yerr=post_std,hatch=patterns[1],edgecolor='k', capsize=7, label='Post')

plt.ylabel('Mean score')

plt.xticks(r + width/2,['SE Challenges','SE Orientations','Self Compassion','Cognitive Reappraisal','Expressive Suppression','Rational Decision Style','Intuitive Decision Style','Empathy'])
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))

plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.3),
          fancybox=True, shadow=False, ncol=5)
plt.ylim([1.5,6])
plt.xticks(rotation=20)


plt.title('Change in scores')
plt.savefig("changes.png",dpi=300)
plt.show()
```
